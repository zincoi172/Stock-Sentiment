```python
import pandas as pd

# Load stock data
stock_df = pd.read_csv("../data/price_data/AAPL.csv", parse_dates=["timestamp"])
stock_df["Date"] = stock_df["timestamp"].dt.date  # Convert to date

# Load sentiment data
sentiment_df = pd.read_csv("../data/sentiment_data/news_apple_sentiment.csv")
sentiment_df["publishedAt"] = pd.to_datetime(sentiment_df["publishedAt"])
sentiment_df["Date"] = sentiment_df["publishedAt"].dt.date

# Aggregate sentiment by day
daily_sentiment = sentiment_df.groupby("Date")["sentiment_score"].mean().reset_index()

# âœ… Merge on same-type Date columns
merged_df = pd.merge(stock_df, daily_sentiment, on="Date", how="left")

# Preview
print(merged_df.columns)
display(merged_df.head())
```
